```python
def save_seasons_ids() -> None:
    seasons_url = "https://api.pubg.com/shards/steam/seasons/"
    header = {
        "Authorization": "Bearer " + PUBG_KEY,
        "Accept": "application/vnd.api+json"
    }

    r = requests.get(seasons_url, headers=header)

    seasons_dict = {
        'seasonType': [],
        'seasonId': [],
        'isCurrentSeason': [],
        'isOffseason': [],
    }

    for season in r.json()['data']:
        seasons_dict['seasonType'].append(season['type'])
        seasons_dict['seasonId'].append(season['id'])
        seasons_dict['isCurrentSeason'].append(season['attributes']['isCurrentSeason'])
        seasons_dict['isOffseason'].append(season['attributes']['isOffseason'])

    seasons_df = pd.DataFrame(seasons_dict)

    seasons_df.to_csv('data/seasons.csv', mode='w', index=False)

save_seasons_ids()
```

### Статистика в пабах

```python

def get_mastery(account_id:str) -> dict:
    def mastery_url(account_id:str) -> str:
        return f"https://api.pubg.com/shards/steam/players/{account_id}/survival_mastery"

    url = mastery_url(account_id)

    header = {
        "Authorization": "Bearer " + PUBG_KEY,
        "Accept": "application/vnd.api+json"
    }

    r = requests.get(url=url, headers=header)

    return r.json()

def get_season_stats(account_id:str, season_id:str):
    def season_stats_url(account_id:str, season_id:str) -> str:
        return f"https://api.pubg.com/shards/steam/players/{account_id}/seasons/{season_id}"

    url = season_stats_url(account_id, season_id)

    header = {
        "Authorization": "Bearer " + PUBG_KEY,
        "Accept": "application/vnd.api+json"
    }

    r = requests.get(url=url, headers=header)

    return r.json()

def get_ranked_season_stats(account_id:str, season_id:str):
    def ranked_season_stats_url(account_id:str, season_id:str) -> str:
        return f"https://api.pubg.com/shards/steam/players/{account_id}/seasons/{season_id}/ranked"

    url = ranked_season_stats_url(account_id, season_id)

    header = {
        "Authorization": "Bearer " + PUBG_KEY,
        "Accept": "application/vnd.api+json"
    }

    r = requests.get(url=url, headers=header)

    return r.json()

def get_player_pub_stats(account_id:str, stats_to_save:list, seasons:list) -> dict:
    # 10 requests total
    mastery_json = get_mastery(account_id)

    general_stats_dict = {
        'playerId': account_id,
        'level': (mastery_json['data']['attributes']['tier'] - 1) * 500 + mastery_json['data']['attributes']['level'],
    }    

    normal_game_mods = ['duo-fpp', 'squad-fpp']
    normal_stats_dict = {stat + '_normal': 0 for stat in stats_to_save}  # for single player
    ranked_stats_dict = {stat + '_ranked': 0 for stat in stats_to_save}  # for single player


    for season_id in seasons:
        normal_stats_json = get_season_stats(account_id, season_id)
        for game_mode in normal_game_mods:
            
            for stat in stats_to_save:
                # print(stat)
                normal_stats_dict[stat + '_normal'] += normal_stats_json['data']['attributes']['gameModeStats'][game_mode][stat]

        ranked_stats_json = get_ranked_season_stats(account_id, season_id)
        for stat in stats_to_save:
            if ranked_stats_json['data']['attributes']['rankedGameModeStats']:
                if 'squad-fpp' in ranked_stats_json['data']['attributes']['rankedGameModeStats']:
                    ranked_stats_dict[stat + '_ranked'] += ranked_stats_json['data']['attributes']['rankedGameModeStats']['squad-fpp'][stat] 
                else:
                    ranked_stats_dict[stat + '_ranked'] += 0

    total_stats = general_stats_dict | normal_stats_dict | ranked_stats_dict

    for matchmaking_mode in ['_normal', '_ranked']:
        for stat in stats_to_save[:-1]:
            cur_stat = stat + matchmaking_mode
            total_stats[cur_stat] = total_stats[cur_stat] / (total_stats['roundsPlayed' + matchmaking_mode] or 999)
    
    return total_stats

def save_players_pub_stats(
            players_ids:list,
            are_pros:bool = False,
            seasons:list = ['division.bro.official.pc-2018-34', 'division.bro.official.pc-2018-33', 'division.bro.official.pc-2018-32'],
        ) -> pd.DataFrame:
    stats_to_save = ['assists', 'dBNOs', 'damageDealt', 'kills', 'revives', 'wins', 'roundsPlayed']

    stats_dict = defaultdict(list)

    for player_id in tqdm(players_ids):
        player_stats = get_player_pub_stats(player_id, stats_to_save, seasons)        
        for stat, value in player_stats.items():
            stats_dict[stat].append(value)

        sleep(60)

    dataframe = pd.DataFrame(stats_dict)
    dataframe['isPro'] = are_pros

    return dataframe

```
